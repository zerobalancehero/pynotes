```python
import matplotlib.pyplot as plt
def arr():
    n= list(x for x in map(int,input("enter the numbers seperated byt space: ").strip().split())if x>=0)
    if not n:
        print ("no input given")
    else:
        plt.plot(n,marker='o')
        plt.title("example table")
        plt.xlabel('x-axis')
        plt.ylabel('y-axis')
        plt.grid()
        plt.show()


arr()
```
